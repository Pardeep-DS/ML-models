```python
#Find the value of Q1, Q2, and IQR for find the outlier
numeric_columns = new_df.select_dtypes(include = ['number']).columns
new_df_numeric = new_df[numeric_columns]

Q1 = new_df_numeric.quantile(0.25)
Q3 = new_df_numeric.quantile(0.75)
IQR = Q3 - Q1
Pos_outlier = Q3 + 1.5 *IQR
Neg_outlier = Q1 - 1.5 * IQR
```

```python
print(Q1)
print('*********************')
print(Q3)
print('*********************')
print(IQR)
print('*********************')
print(Pos_outlier)
print('*********************')
print(Neg_outlier)
```

```python
#Handel the outlier
filtered_data = new_df_numeric[~((new_df_numeric < Neg_outlier) | (new_df_numeric > Pos_outlier)).any(axis = 1)]
non_numeric_columns = new_df.select_dtypes(exclude = ['number']).columns
new_df_filtered = pd.concat([filtered_data, new_df[non_numeric_columns]], axis = 1).loc[filtered_data.index]

new_df = new_df_filtered
```
