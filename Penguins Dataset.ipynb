```python
#Pass the independent variables to x and dependent variable to y
x = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
#To check the first five record's of the independent variables
x.head()
```

```python
#To see the first five records of the dependent variable
y.head()
```

```python
#Perform feature scalling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_sc = sc.fit_transform(x)
```

```python
x_sc
```

```python
#To see the data after feature scalling
pd.DataFrame(x_sc)
```

```python
#To split the data into train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_sc, y, test_size = 0.2, random_state = 101)
```

```python
#Build the decision tree model
from sklearn.tree import DecisionTreeClassifier
dt1 = DecisionTreeClassifier(criterion = 'gini')
dt1.fit(x_train, y_train)
dt2 = DecisionTreeClassifier(criterion = 'entropy')
dt2.fit(x_train, y_train)
```

```python
#Predict the values by the model
y_predict_dt1_train = dt1.predict(x_train)
y_predict_dt1_test = dt1.predict(x_test)

y_predict_dt2_train = dt2.predict(x_train)
y_predict_dt2_test = dt2.predict(x_test)
```

```python
#Import some libraries to check the accuracy
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
#Make confusion matrix
confusion_matrix(y_test, y_predict_dt1_test)
```

```python
#Evaluating Decision Tree Model
print(classification_report(y_train, y_predict_dt1_train))
```

```python
#Evaluating Decision Tree Model  
print(classification_report(y_test, y_predict_dt1_test))
```

```python
#To check the accuracy of train and test by dt1
print("Train Accuracy :", accuracy_score(y_train, y_predict_dt1_train))
print("*********"*5)
print("Test Accuracy :", accuracy_score(y_test, y_predict_dt1_test))
```

```python
#To check the accuracy of train and test by dt2
print("Train Accuracy :", accuracy_score(y_train, y_predict_dt2_train))
print("*********"*5)
print("Test Accuracy :", accuracy_score(y_test, y_predict_dt2_test))
```

```python
#Import different - different models
from mlxtend.classifier import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
#Train the different - different models on the same dataset
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()
clf3 = KNeighborsClassifier()
clf4 = GaussianNB()
clf5 = RandomForestClassifier()
sclf = StackingClassifier(classifiers = [clf2, clf3, clf4, clf5], meta_classifier = clf1)

print("5 fol corss - validations \n")

for clf, label in zip([clf1, clf2, clf3, clf4, clf5, sclf], ["Decision Tree", "kNeghbors Classifier",
                                                            "Naive Bayes", "Random Forest", "Stacking Classifier"]):
    scores = cross_val_score(clf,x_train,y_train,cv=5,scoring='accuracy')
    print("Accuracy : %0.2f (+/-%0.2f)[%s]"  %(scores.mean(), scores.std(), label))
```
