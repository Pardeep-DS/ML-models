```python
#Split the data into independent and dependent variable
x = df.iloc[:,:4]       #Independent variables
y = df.iloc[:,-1:]      #Dependent variables 
```

```python
#See the first five records of the independent variable's
x.head()
```

```python
#See the last five records of the dependent variable 
y.head()
```

```python
#Perform standardization 
from sklearn.preprocessing import StandardScaler
Std_scaler = StandardScaler()
x_sc = Std_scaler.fit_transform(x)
x_sc
```

```python
#Split the data into train and test 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_sc, y, test_size = 0.2, random_state = 101)
```

```python
#See the training data of independent variable's
x_train
```

```python
#See the training data of dependent variable 
y_train
```

```python
#Build the model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
#Predict the x_test data with the building model to check the accuracy of it 
y_pred = model.predict(x_test)
```

```python
#Import r2_score for calculating the accuracy 
from sklearn.metrics import r2_score
```

```python
#Print the accuracy of the model 
print("Accuracy is:", r2_score(y_test, y_pred))
```
