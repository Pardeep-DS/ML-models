```python
#Encoding of 'Gender' variable
df['Gender'] = df['Gender'].astype('category')
df['Gender'] = df['Gender'].cat.codes
```

```python
#Encoding of 'Have you ever had suicidal thoughts ?' variable
df['Have you ever had suicidal thoughts ?'] = df['Have you ever had suicidal thoughts ?'].astype('category')
df['Have you ever had suicidal thoughts ?'] = df['Have you ever had suicidal thoughts ?'].cat.codes
```

```python
#Encoding of 'Dietary Habits' variable
df['Dietary Habits'] = df['Dietary Habits'].astype('category')
df['Dietary Habits'] = df['Dietary Habits'].cat.codes
```

```python
#Check the first five record after encoding
df.head()
```

```python
#Check the last five record after encoding
df.tail()
```

```python
#Pass the independent variable to x and dependent variable to y
x = df[['Gender', 'Age', 'Academic Pressure', 'Study Satisfaction', 'Sleep Duration', 'Dietary Habits', 'Have you ever had suicidal thoughts ?', 'Study Hours', 'Financial Stress', 'Family History of Mental Illness']]
y = df[['Depression']]
```

```python
#Perform one hot encoding of sleep duration
from sklearn.preprocessing import OneHotEncoder
One_Hot_Encoder = OneHotEncoder()
new_x = One_Hot_Encoder.fit_transform(x)
x = new_x
```

```python
#check the first five records of the dependent variable
y.head()
```

```python
#check the last five records of the dependent variable
y.tail()
```

```python
#import libraries for spliting data into train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 200)
```

```python
#check the number of row's and column's
print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
```

```python
#import libraries for build the logistic model
from sklearn.linear_model import LogisticRegression
linear_model = LogisticRegression()
linear_model.fit(x_train,y_train)
```

```python
#find the value of y predict
y_predict = linear_model.predict(x_test)
```

```python
#import library for make confusion matrix 
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
cm = confusion_matrix(y_predict, y_test)
```

```python
#To Display the confusion matrix 
disp = ConfusionMatrixDisplay(cm, display_labels=[0,1])
disp.plot(cmap = 'Blues')
plt.show()
```

```python
#print confusion matrix
confusion_matrix = confusion_matrix(y_test, y_predict)
print(confusion_matrix)
```
