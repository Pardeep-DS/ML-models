```python
#To Calculate the value of Q1, Q3 and IQR
numeric_column = df.select_dtypes(include = ['number']).columns
df_numeric = df[numeric_column]

Q1 = df_numeric.quantile(0.25)
Q3 = df_numeric.quantile(0.75)

IQR = Q3 - Q1

Positive_outlier = Q3 + 1.5 * IQR
Negative_outlier = Q1 - 1.5 * IQR
```

```python
#To Print the value's of Q1, Q2, and IQR
print(Q1)
print('****************')
print(Q3)
print('****************')
print(IQR)
print('****************')
print(Positive_outlier)
print('****************')
print(Negative_outlier)
```

```python
#To handel the outlier
filtered_data = df_numeric[~((df_numeric < Negative_outlier) | (df_numeric > Positive_outlier)).any(axis = 1)]
non_numeric_column = df.select_dtypes( exclude = ['number'] ).columns

df_filtered_data = pd.concat([filtered_data, df[non_numeric_column]], axis = 1).loc[filtered_data.index]
new_df = df_filtered_data                
```

```python
#To check the duplicate records
new_df.duplicated().sum()
```

```python
#Pass the independent variables to the x and the dependent variable to y
x = new_df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']]
y = new_df[['Outcome']]
```

```python
#To check the first five record of Independent variables
x.head()
```

```python
#Split the data into train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 101)
```

```python
#To check the train records of independent variables
x_train
```

```python
#To check the train records of dependent variable
y_train
```

```python
#import the model from sklearn
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
#Train the model on x_train and y_train data
model.fit(x_train, y_train)
```

```python
#Predict the result of x_test data
y_predict = model.predict(x_test)
```

```python
#import accuracy_score to calculate the accuracy of the model
from sklearn.metrics import accuracy_score
```

```python
#To print the accuracy of the model
print(accuracy_score(y_test, y_predict))
```

```python
#import cross_val_score library for improve the accuracy
from sklearn.model_selection import cross_val_score
```

```python
#Cross_val_score provide the different - different accuracy on the behalf of cv
accuracy = cross_val_score(model, x_test, y_test, cv = 10)
accuracy
```

```python
#Pick the best accuracy from the above
print('The best accuracy after 10 split is:', accuracy[2])
```
